```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im=cv2.imread("b.png")
im=cv2.resize(im,(100,100))
plt.imshow(im)
```

```python
def bound_fill(x,y,fill,bound):
    if np.array_equal(im[x,y],fill)== False and  np.array_equal(im[x,y],bound)== False:
        im[x,y]=fill
        bound_fill(x,y+1,fill,bound)
        bound_fill(x,y-1,fill,bound)
        bound_fill(x+1,y,fill,bound)
        bound_fill(x-1,y,fill,bound)
```

```python
bound_fill(40,40,(255,255,255),(0,0,0))
plt.imshow(im)
```
