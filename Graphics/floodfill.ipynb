```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im=cv2.imread("b.png")
im=cv2.resize(im,(100,100))
plt.imshow(im)
```

```python
def flood_fill(x,y,old,new):
    if np.array_equal(im[x,y],old)== True:
        im[x,y]=new
        flood_fill(x,y+1,old,new)
        flood_fill(x,y-1,old,new)
        flood_fill(x+1,y,old,new)
        flood_fill(x-1,y,old,new)
```

```python
print(im[40,40])
flood_fill(40,40,(183,58,103),(200,5,6))
plt.imshow(im)
```
