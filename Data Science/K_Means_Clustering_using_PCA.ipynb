```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
scale_data = scaler.transform(df)
```

```python
scale_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scale_data)
```

```python
x_pca = pca.transform(scale_data)
```

```python
scale_data.shape
```

```python
x_pca.shape
```

```python
scale_data
```

```python
x_pca
```

```python
plt.figure(figsize=(7,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=iris['target'])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.show()
```

```python
dataset=pd.DataFrame({
    'PCA_1':x_pca[:,0],
    'PCA_2':x_pca[:,1],
    'cluster': iris.target
        
    }
)

print(dataset)
```

```python
centroids={}

for i in range(3):
  result=[]
  result.append(dataset.loc[dataset['cluster']==i]['PCA_1'].mean())
  result.append(dataset.loc[dataset['cluster']==i]['PCA_2'].mean())

  centroids[i]=result
```

```python
print(centroids)
```

```python
plt.figure(figsize=(7,6))
plt.scatter(dataset['PCA_1'],dataset['PCA_2'],c=iris.target)
plt.xlabel("PCA_1")
plt.ylabel("PCA_2")
plt.show()
```

```python
def closest_distance_from_centroid(dataset,centroids):
    for i in range(3):
      dataset['dist_from_{}'.format(i)]=(
          np.sqrt(
                (dataset['PCA_1']-centroids[i][0])**2
              +(dataset['PCA_2']-centroids[i][1])**2
                ))
    
    for j in range(150):
      min=float('inf')
      index=-1
      for i in range(3):
        k=dataset.at[j,'dist_from_{}'.format(i)]
        if min>k:
          min=k
          index=i  
      dataset.at[j,'closest']=index
    return dataset 
```

```python
dataset=closest_distance_from_centroid(dataset,centroids)
```

```python
dataset
```

```python
def update_centroids(k):
  for i in range(3):
    centroids[i][0] = np.mean(dataset[dataset['closest'] == i]['PCA_1'])
    centroids[i][1] = np.mean(dataset[dataset['closest'] == i]['PCA_2'])
    
  return k
```

```python
centroids=update_centroids(centroids)
print(centroids)
```

```python
while True:
  l=dataset['closest'].copy(deep=True)
  centroids=update_centroids(centroids)
  dataset=closest_distance_from_centroid(dataset,centroids)
  if l.equals(dataset['closest']):
    break
```

```python
plt.figure(figsize=(7,6))
plt.scatter(dataset['PCA_1'],dataset['PCA_2'],c=dataset.closest)
plt.xlabel("PCA_1")
plt.ylabel("PCA_2")
plt.show()
```
